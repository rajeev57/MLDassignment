```python
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline
```

We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1. fitting it to the periodic function
2. plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
temp_max=np.array([39,41,43,47,49,51,45,38,37,29,27,25])
temp_min=np.array([21,23,27,28,32,35,31,28,21,19,17,18])
```

```python
#not periodic 
month = np.arange(1,13)
fig= plt.figure(figsize=(10,10))
plt.plot(month,temp_max,'o',color='red',label="temp_max")
plt.plot(month,temp_min,'o',color='green',label="temp_min")
plt.legend()
plt.show()
```

```python
from scipy.optimize import curve_fit
```

```python
def func1(x,a,b,c):
    return (a+b*np.cos((x+c) * 1.8 * np.pi / x.max()))
```

```python
opt1,pcov = curve_fit(func1,month,temp_max,[40,20,0])
opt2,pcov = curve_fit(func1,month,temp_min,[-40,20,0])
print("optimum1 ",opt1," optimum2 ",opt2)
```

```python
days = np.linspace(0, 12, num=365)

plt.figure(figsize=(10,10))
plt.plot(month, temp_max, 'o',color='red',label="temp_max")
plt.plot(days, func1(days, *opt1), '--',color='green',label="temp_max curve")
plt.plot(month, temp_min, 'o',color='green',label="temp_min")
plt.plot(days, func1(days, *opt2), '--',color='red',label="temp_min curve")
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.legend()
plt.show()
```

```python
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
```

```python
titanic=pd.read_csv(url,skip_blank_lines=True)
```

```python
last_row=len(titanic)
titanic=titanic.drop(titanic.index[last_row-1])
df=titanic['sex']
df1=df.value_counts()
df1
```

```python
fig = plt.figure(figsize=(10,10))
labels='male','female'
explode = (0, 0.05)
colors = ['green', 'red']
plt.pie(df1.values.tolist(),labels=labels,shadow=True,explode=explode,colors=colors,autopct='%1.1f%%',startangle=90)
plt.title('male/female proportion')
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
colors = {'male':'blue', 'female':'orange'}
plt.scatter(titanic.fare,titanic.age,color=df.apply(lambda x: colors[x]))
male = mpatches.Patch(color='blue', label='male')
female = mpatches.Patch(color='orange', label='female')
plt.title('Fare paid and the Age')
plt.legend(handles=[male,female])
plt.show()
```
